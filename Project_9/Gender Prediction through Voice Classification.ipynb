#**Gender Prediction through Voice**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#reading the dataset
df = pd.read_csv('/content/voice.csv')

df.head()
```

```python
df.info()
```

```python
df['label'].value_counts()
#balance class
```

```python
x = df.iloc[:,:-1].values
y = df['label'].values
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

#normalisation
scaler = StandardScaler()
x = scaler.fit_transform(x)

#encoding
label = LabelEncoder()
y = label.fit_transform(y)
```

```python
x[:5] #an array and normalized
```

```python
y #labelled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
x_train.shape
```

```python
#modeling
#SVM Kernelling - Linear, Polynomial, RBF(Radial Based Function), Sigmoid
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

SVM_bank = SVC() #default is RBF

#training

SVM_bank.fit(x_train, y_train)

#prediction

y_hat = SVM_bank.predict(x_test)

#evaluation

print('The accuracy based on RBF kernel is:', (accuracy_score(y_test, y_hat))*100)
```

```python
#modeling
#SVM Kernelling - Linear, Polynomial, RBF(Radial Based Function), Sigmoid
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

SVM_bank = SVC(kernel = 'linear')

#training

SVM_bank.fit(x_train, y_train)

#prediction

y_hat = SVM_bank.predict(x_test)

#evaluation

print('The accuracy based on Linear kernel is:', (accuracy_score(y_test, y_hat))*100)
```

```python
#modeling
#SVM Kernelling - Linear, Polynomial, RBF(Radial Based Function), Sigmoid
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

SVM_bank = SVC(kernel = 'poly')

#training

SVM_bank.fit(x_train, y_train)

#prediction

y_hat = SVM_bank.predict(x_test)

#evaluation

print('The accuracy based on Poly kernel is:', (accuracy_score(y_test, y_hat))*100)
```

```python
#modeling
#SVM Kernelling - Linear, Polynomial, RBF(Radial Based Function), Sigmoid
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

SVM_bank = SVC(kernel = 'sigmoid')

#training

SVM_bank.fit(x_train, y_train)

#prediction

y_hat = SVM_bank.predict(x_test)

#evaluation

print('The accuracy based on Sigmoid kernel is:', (accuracy_score(y_test, y_hat))*100)
```

```python
#cross validation - mean value of accuracy

from sklearn.model_selection import cross_val_score
import numpy as np

SVM_bank = SVC()

mean_accuracy = cross_val_score (SVM_bank, x, y, cv = 5, scoring = 'accuracy', verbose = 2)
# verbose = for deep learning (1 = only show the process)
# cv = how many times to run the dataset randomly (shuffling randomly)


print('The total number of accuracy are:', mean_accuracy)
print('')
print('The average of all is:', (np.mean(mean_accuracy))*100)
```

```python
#cross validation for SVM linear kernel
SVM_bank = SVC(kernel = 'linear')

mean_accuracy = cross_val_score (SVM_bank, x, y, cv = 20, scoring = 'accuracy', verbose = 2)

print('The total number of accuracy are:', mean_accuracy)
print('')
print('The average of all is:', np.mean((mean_accuracy))*100)
```

```python
#cross validation for SVM polynomial kernel
SVM_bank = SVC(kernel = 'poly')

mean_accuracy = cross_val_score (SVM_bank, x, y, cv = 5, scoring = 'accuracy', verbose = 2)

print('The total number of accuracy are:', mean_accuracy)
print('')
print('The average of all is:', (np.mean(mean_accuracy))*100)
```

```python
#cross validation for SVM sigmoid kernel
SVM_bank = SVC(kernel = 'sigmoid')

mean_accuracy = cross_val_score (SVM_bank, x, y, cv = 5, scoring = 'accuracy', verbose = 2)
#verbose = for deep learning (1 = only show the process)
# cv = how many times to run the dataset randomly (shuffling randomly)


print('The total number of accuracy are:', mean_accuracy)
print('')
print('The average of all is:', (np.mean(mean_accuracy))*100)
```

###**Prediction of Voice using KNN Algorithm**

###**Prediction of Voice using Decision Tree Algorithm**

```python
df.head()
```

```python
x2 = df.iloc[:,:-1].values
y2 = df['label'].values
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

label = LabelEncoder()

y2 = label.fit_transform(y2)

#NORMALISATION

scaler = StandardScaler()

x2 = scaler.fit_transform(x2)
```

```python
#splitting dataset for training and testing
from sklearn.model_selection import train_test_split

x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, train_size = 0.8, random_state = 24 )

print(x2_train.shape)
print(x2_test.shape)
```

```python
#cross validation for Decision Tree Algorithm

tree = DecisionTreeClassifier()

mean_accuracy = cross_val_score (tree, x2, y2, cv = 5, scoring = 'accuracy', verbose = 2)

print('The total accuracy score is:', mean_accuracy)
print('')
print('The mean accuracy score is:', (np.mean(mean_accuracy))*100)
```

###**Prediction of Voice using Logistic Regression Algorithm**

```python
x3 = df.iloc[:,:-1].values
y3 = df['label'].values
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

#labeling target value

label = LabelEncoder()
y3 = label.fit_transform(y3)

#normalisation

scaler = StandardScaler()
x3 = scaler.fit_transform(x3)
```

```python
from sklearn.model_selection import train_test_split

x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y3, train_size = 0.8, random_state = 24)

print(x3_train.shape)
print(x3_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#training
LR = LogisticRegression()
LR.fit(x3_train, y3_train)

#prediction
y3_hat = LR.predict(x3_test)

#evaluation
print('The accuracy score is:', (accuracy_score(y3_test, y3_hat))*100)
```

```python
#cross validation

LR = LogisticRegression()

mean_accuracy = cross_val_score(LR, x3, y3, cv = 5, scoring = 'accuracy', verbose = 2)

print('The total accuracy score is:', mean_accuracy)
print('The mean of all accuracy is:', (np.mean(mean_accuracy))*100)
```
